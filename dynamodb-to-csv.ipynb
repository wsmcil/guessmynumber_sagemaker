```python
import boto3
from boto3.dynamodb.conditions import Key, Attr
import decimal
import datetime
import pandas as pd
import json
```

```python
# Helper class to convert a DynamoDB item to JSON.
class DecimalEncoder(json.JSONEncoder):
    def default(self, o):
        if isinstance(o, decimal.Decimal):
            if o % 1 > 0:
                return float(o)
            else:
                return int(o)
        return super(DecimalEncoder, self).default(o)
```

```python
dynamodb = boto3.resource('dynamodb', region_name='us-west-2', endpoint_url="https://dynamodb.us-west-2.amazonaws.com")
table = dynamodb.Table('ml-game-observations')
```

```python
pe = "info"

response = table.scan(
    ProjectionExpression=pe
    )
```

```python
print(response)
```

```python
print (response['Items'][0]['info']['actual'])
```

```python
df = pd.DataFrame(columns = ['actual', 'guess', 'algorithm', 'image'])
row=0

for i in response['Items']:
        #print(json.dumps(i, cls=DecimalEncoder))
        #print(i['info']['actual'])
        df.loc[row]=[i['info']['actual'], i['info']['guess'], i['info']['algorithm'], i['info']['image']]
        row=row+1
```

```python
df
```

```python
df['actual'] = df['actual'].str[0]
df['algorithm'] = df['algorithm'].str[0]
df['guess'] = df['guess'].str[0]
```
